```python
#5. Write a function with one parameter, lg. The function must return the number of words (alphanumeric strings) that have the length lg. Call the function and print the returned result.

def count_words(lg):
  return len([word for word in text2.tokens if len(word) == lg and word.isalnum()])

print(count_words(5))
```

```python
#6. Write a function (with one parameter, lt) that prints the first N distinct words (from the alphabetically ordered vocabulary) that begin with the letter lt. Call the function.
def first_n_words(st, sz):
  lg = 0
  sorted_text = sorted(text2.tokens)
  word_list = [word for word in sorted_text if word.startswith(st)]
  word_list = list(dict.fromkeys(word_list))
  return word_list[:sz]

print(first_n_words("t", 5))
```

```python
#7. Print the shortest and longest words (alphanumeric strings). If there are multiple words with minimum or maximul length, you must print all of them.


al_num_text = [word for word in text2.tokens if word.isalnum()]
shortest_len = min(len(word) for word in al_num_text)
print (shortest_len)
shortest_words = [word for word in al_num_text if len(word) is shortest_len]
print (shortest_words)

maximum_len = max(len(word) for word in al_num_text)
print (maximum_len)
maximum_words = [word for word in al_num_text if len(word) is maximum_len]
print (maximum_words)


# mins = min(len(word) for word in line.split())
```

```python
#8. Print the first N most frequent words (alphanumeric strings) together with their number of appearances.
from collections import Counter 
al_num_text = [word for word in text2.tokens if word.isalnum()]
Counter = Counter(al_num_text) 
def most_fequent_words(lg): 
  most_occur = Counter.most_common(lg) 
  return most_occur
  
print(most_fequent_words(10)) 
```
