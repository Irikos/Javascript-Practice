```python
import pymc as pm
import numpy as np

from sklearn.datasets import make_spd_matrix
from scipy.special import softmax    
```

**In this notebook, we will be looking into Topic Modeling with PyMC2.**

We will be implementing a *Latent Dirichlet Allocation* (LDA) model and we'll be exploring some methods for getting more informations about the topic structure of documents.

We will also implement some variants of LDA, namely the *Correlated Topic Model* (CTM) and the *Dynamic Topic Model* (DTM).

We will be implementing all of our models as described in http://www.cs.columbia.edu/~blei/papers/BleiLafferty2009.pdf .

First, let's make some classes for our vocabularies:

```python
#Todo: make a single class out of them
class Vocabulary(object):
    """Gets an array of docs and tokenizes each word in the corpus.
    """
    def __init__(self, docs):
        self.word_ids = {}
        self.ids_counter = 0
        for doc in docs:
            for word in doc:
                if word not in self.word_ids:
                    self.word_ids[word] = self.ids_counter
                    self.ids_counter = self.ids_counter + 1

    def __getitem__(self, idx):
        try:
            x = self.word_ids[idx]
        except KeyError:
            x = self.ids_counter # last id is UNK
        return x
        
    def __len__(self):
        return self.ids_counter
    
class Vocabulary_Time(object):
    """Gets an array of time-slices of docs and tokenizes each word in the corpus.
    Will be used for the Dynamic Topic Model.
    """
    def __init__(self, docs_ts):
        self.word_ids = {}
        self.ids_counter = 0
        for docs in docs_ts:
            for doc in docs:
                for word in doc:
                    if word not in self.word_ids:
                        self.word_ids[word] = self.ids_counter
                        self.ids_counter = self.ids_counter + 1

    def __getitem__(self, idx):
        try:
            x = self.word_ids[idx]
        except KeyError:
            x = self.ids_counter # last id is UNK
        return x
    
    def __len__(self):
        return self.ids_counter
```

Let's settle on some variable names and let's describe the generative process.

Let $M$ denote the number of documents.

Let $N_i$ denote the number of words in a given document $i$.

$\mathbf{\alpha}$ is the parameter of the Dirichlet prior on the per-document topic distributions.

$\mathbf{\beta}$ is the parameter of the Dirichlet prior on the per-topic word distributions.

$\mathbf{\theta_i}$ will be the topic distribution over document $i$.

$\mathbf{\phi_k}$ is the word distribution for topic k.

$\mathbf{z_{i,j}}$ is the topic for the $j$-th word in document $i$.

$\mathbf{w_{i,j}}$ is the specific word.

**Generative process:**

1. Draw $\mathbb{\theta_{i}} \sim Dirichlet(\alpha), i \in \{1,...,M\}$.

2. Draw $\mathbb{\phi_{k}} \sim Dirichlet(\beta), k \in \{1,...,N_i\}$.

3. For $i \in \{1,...,M\}$ and $j \in \{1,...,N_i\}$ draw:

    3.1. a topic $z_i \sim Categorical(\theta_i)$.

    3.2. a word $w_{i,j}$ $\sim$ $Categorical$ $(\phi_{z_{i,j}})$.



(Fig. 1) Plate notation for the LDA model.

We'll be using the Markov Chain Monte Carlo method from PyMC2 to find the approximate posterior distribution.

*Note: PyMC's MCMC is using BinaryMetropolis by default.*

Define the LDA model class:

```python
class LDA(object):
    """The LDA model class.
    The class initializer takes a list of documents.
    K is the number of topics in the corpus. Default = 2.
    alpha_scale and beta_scale are hyperparameters for the Dirichlet parameters scale. Default = 1.
    """
    def __init__(self, docs, K=3, alpha_scale=1, beta_scale=1):
        self.gen_vocab(docs)
        
        self.K = K
        
        self.alpha = np.ones(self.K) * alpha_scale
        self.beta = np.ones(self.V) * beta_scale
        
        self._init_LDA()
        
    def _init_LDA(self):
        self.theta = pm.Container(
            [
                pm.CompletedDirichlet(
                    "theta_%s" % i, 
                    pm.Dirichlet("ptheta_%s" % i, theta = self.alpha)) 
                for i in range(self.D)
            ]
        )
        
        self.phi = pm.Container(
            [
                pm.CompletedDirichlet(
                    "phi_%s" % j, 
                    pm.Dirichlet("pphi_%s" % j, theta = self.beta)) 
                for j in range(self.K)
            ]
        )
        
        self.z = pm.Container(
            [
                pm.Categorical(
                    "z_%i" % d,
                    p = self.theta[d],
                    size = self.Nd[d]
                ) 
                for d in range(self.D)
            ]
        )
        
        self.w = pm.Container(
            [
                pm.Categorical(
                    "w_%i_%i" % (d,i), #document d, word i
                    p = pm.Lambda(
                        "phi_z_%i_%i" % (d,i),
                        lambda z = self.z[d][i], phi = self.phi : phi[z]
                    ),
                    value = self.collection[d][i],
                    observed = True
                ) 
                for d in range(self.D) for i in range(self.Nd[d])
            ]
        )       
        #In pymc, this means that term will be a likelihood term using the value argument as the observed value, 
        #for purposes of calculating the gradient of the log likelihood when doing MCMC proposals. 
        
        self.model = pm.Model([self.theta, self.phi, self.z, self.w])
        self.mcmc = pm.MCMC(self.model)
#         self.mcmc = pm.MCMC([self.theta, self.phi, self.z, self.w])
        
    def gen_vocab(self, docs):
        self.vocabulary = Vocabulary(docs)
        self.collection = [[self.vocabulary[w] for w in d] for d in docs]
        self.V = len(self.vocabulary)
        self.D = len(self.collection)
        self.Nd = [len(doc) for doc in self.collection]
        
    def get_parameters(self):
        return (self.V, self.D, self.alpha, self.beta, self.Nd)
    
    def print_parameters(self):
        print('V: %s \nD: %s \nalpha: %s \nbeta: %s \nNd: %s' % 
             self.get_parameters())        

#     def fit(self, iter=40000, burn=15000, thin=10, tune_interval=2, tune_throughout=False):
    def fit(self, iter=10000, burn=5000):
        self.mcmc.sample(iter, burn)
    
    def trace_var(self, var_name, idx = None):
        return self.mcmc.trace(var_name)
    
    def raftery_lewis(self, q=0.025, r=0.01):
        pm.raftery_lewis(self.mcmc, q, r)
        
    def get_topic_distribution(self): # over words
        return self.phi.value
    
    def get_word_distribution(self): # in documents over topics
        return self.w.value
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import scipy.stats as stats
import seaborn as sns

def plot_words_per_topic(model, ax=None):
    if ax is None:
        plt.clf()
        fig, ax = plt.subplots(1,1)
    words = model.z.value
    topic_dist = dict()
    for k_i in words:
        for k in k_i:
            if k not in topic_dist:
                topic_dist[k] = 0
            topic_dist[k] += 1
    ax.bar(topic_dist.keys(), topic_dist.values())
    ax.set_xlabel("Topics")
    ax.set_ylabel("Counts")
    ax.set_title("Document words per topics")
    plt.show()
    
def plot_word_dist(model, ax=None):
    topics = model.phi.value
    if ax is None:
        plt.clf()
        from math import ceil
        fig, ax = plt.subplots( ceil(len(topics)/2) , 2, figsize=(10,10))
    for i, t in enumerate(topics):
        ax[i//2][i%2].bar(range(len(t[0])), t[0])
        ax[i//2][i%2].set_title("Topic %s" % i)
        ax[i//2][i%2].set(xlabel='word index', ylabel='word probability')
    plt.suptitle("Vocab word proportions per topic")
    fig.subplots_adjust(hspace=0.5, wspace=0.5)
```

**Calculating document-similarity using the posterior topic proportions.**

We can define a topic-based similarity measure between documents. One metric that can be used between the posteriors of two documents topic proportions can be the $Hellinger$ $distance$, or the $KL-Divergence$ between the posteriors.

Let $f$ and $d$ be two documents and let $\hat{\theta}_{d,k}$ be the posterior topic proportions estimator for topic $k$ in the $d^{th}$ document.

Let the $Hellinger$ $distance$ be defined as:

$hellinger(f, d) := \sum_{k=1}^K(\sqrt{\hat{\theta}_{f,k}} - \sqrt{\hat{\theta}_{d,k}})^2$

And let the $KL-Divergence$ be defined as:

$D_{KL}(f, d) := \sum_{k=1}^K\theta_{f,k}log(\frac{\hat{\theta}_{f,k}}{\hat{\theta}_{d,k}})$.

Note that the $KL-Divergence$ is not symmetric.

Let's define the functions and test our metrics.

```python
def plot_Hellinger_similarity_matrix(model):
    doc_topics = []
    for d in range(model.D):
        doc_topics.append(model.trace_var('theta_%i' % d)[-1])
    
    similarity_matrix = np.zeros( (model.D, model.D) )
    
    for i, doc_i in enumerate(doc_topics):
        for j, doc_j in enumerate(doc_topics):
            similarity_matrix[i][j] = np.sum(np.sqrt(doc_i) - np.sqrt(doc_j))
    
    ax = sns.heatmap(similarity_matrix, annot=True, linewidths=.5).set_title('Hellinger similarity matrix between documents (i, j).')
    #This fixes cropped top and bottom problem
    b, t = plt.ylim()
    b += 0.5
    t -= 0.5
    plt.ylim(b, t)
    
    plt.show()
    
    return similarity_matrix

_ = plot_Hellinger_similarity_matrix(model)
```

```python
def plot_KLD_matrix(model):
    doc_topics = []
    for d in range(model.D):
        doc_topics.append(model.trace_var('theta_%i' % d)[-1])
    
    KLD_matrix_i = np.zeros( (model.D, model.D) )
    
    for i, doc_i in enumerate(doc_topics):
        for j, doc_j in enumerate(doc_topics):
            KLD_matrix_i[i][j] = np.sum(np.multiply(doc_i, np.log(np.divide(doc_i, doc_j))))
    
    ax = sns.heatmap(KLD_matrix_i, annot=True, linewidths=.5).set_title('KLD(doc_i || doc_j) matrix.')
    #This fixes cropped top and bottom problem
    b, t = plt.ylim()
    b += 0.5
    t -= 0.5
    plt.ylim(b, t)
    
    plt.show()
    
    return KLD_matrix_i

_ = plot_KLD_matrix(model)
```

**Question: What about new documents? How do we assign topics to it?**

The best solution in this situation is to run inference on a new corpus of documents that contains the old corpus plus the new documents.

A faster method would be running the MCMC algorithm again on the new corpus, keeping the old posterior distributions and topic assignments, and using these as the new priors. 

The convergence of the Markov Chain should happen faster and the chain will eventually converge to the (new) stationary distribution, due to the **ergodicity** property of the time-homogeneous Markov Chain that is produced by PyMC's Metropolis sampler. **Ergo** (pun intended), this solution should be viable.

We'll do the first method.

The second approach to topic modelling is the **Correlated Topic Model**. The CTM is a variant for the LDA model that takes into account the correlation between the occurrence of topics. In most text corpora, it is natural to expect that the occurrences of the underlying latent topics will be highly correlated \[[1](http://www.cs.columbia.edu/~blei/papers/BleiLafferty2009.pdf)\].

Let's describe the generative process of CTM:

Let $f(\eta_i) = \frac{\mathcal{e}^{\eta_i}}{\sum_j \mathcal{e}^{\eta_j}}$, the $softmax$ function.

**Generative process:**

1. Draw $\mathbb{\eta}$ $\sim$ $\mathcal{N}$ $(\mu, \Sigma)$.

2. For $n \in \{1,...,N\}$ draw:

    2.1. a topic assignment $z_n \sim Categorical(f(\eta))$.

    2.2. a word $w_{i,j}$ $\sim$ $Categorical$ $(\beta_{z_{n}})$.
    
The generative process is identical with the LDA model, except that the topic proportions are drawn from a logistic Gaussian distribution rather than a Dirichlet distribution.



(Fig. 2) Plate notation for the CTM.

Define the CTM class:

```python
class CTM(object):
    
    def __init__(self, docs, K=2, alpha_scale=1, beta_scale=1):
        self.vocabulary = Vocabulary(docs)
        self.collection = [[self.vocabulary[w] for w in d] for d in docs]
        self.K = K
        self.V = len(self.vocabulary)
        self.D = len(self.collection)
        
        self.alpha = np.ones(self.K) * alpha_scale
        self.beta = np.ones(self.V) * beta_scale
        
        self.Nd = [len(doc) for doc in self.collection]
        
        self.mu = np.ones(self.K)
        from sklearn.datasets import make_spd_matrix
        self.cov = make_spd_matrix(K) # Covariance matrix needs to be positive-definite matrix
                                      # PDF exists only when covariance matrix is positive-definite
        self._init_CTM()
        
    def _init_CTM(self):
        self.eta = pm.MvNormal('eta', mu=self.mu, tau=self.cov)
        
        from scipy.special import softmax
        self.z = pm.Container(
            [
                pm.Categorical(
                "z_%i" % d,
                    p = softmax(self.eta.value),
                    size = self.Nd[d]
                )
                for d in range(self.D)
            ]
        )
        
        self.phi = pm.Container(
            [
                pm.CompletedDirichlet(
                    "phi_%s" % j, 
                    pm.Dirichlet("pphi_%s" % j, theta = self.beta)) 
                for j in range(self.K)
            ]
        )
        
        self.w = pm.Container(
            [
                pm.Categorical(
                    "w_%i_%i" % (d,i), #documentul d, cuvantul i
                    p = pm.Lambda(
                        "phi_z_%i_%i" % (d,i),
                        lambda z = self.z[d][i], phi = self.phi : phi[z]
                    ),
                    value = self.collection[d][i],
                    observed = True
                ) 
                for d in range(self.D) for i in range(self.Nd[d])
            ]
        )       
      
        self.model = pm.Model([self.eta, self.phi, self.z, self.w])
        self.mcmc = pm.MCMC(self.model)

    def fit(self, iter=10000, burn=5000):
        self.mcmc.sample(iter, burn)
```

```python
CTM_Topic = CTM(corpus) 
CTM_Topic.fit(1000, 500)
```

The final topic we'll be using is the Dynamic Topic Model. The DTM captures the evolution of topics over time, assuming the document corpus is sequentially organized. We will model the documents of each slice with a K-component topic model, where the topics associated with slice $t$ evolve from the topics associated with slice $t-1$.

We will describe the generative process of the DTM.

Let $\pi_{t,k}$ denote a multivariate Gaussian random variable for the topic $k$ in slice $t$ and let $T$ be the number of time slices. We will chain ${\pi_{1,k}, ... , \pi_{T,k}}$  to take the evolution of the corpus into account. The function $f$ is the same $softmax$ function defined at the CTM section.

**Generative process:**

1. Draw the topics $\pi_t$ $\sim$ $\mathcal{N}$ $(\pi_{t-1}, \Sigma I)$.

2. For each document:

    2.1. Draw $\theta_d \sim Dirichlet(\alpha)$.

    2.2 For each word:
    
      2.2.1 Draw $Z\sim Categorical (\theta_d)$.
     
      2.2.2 Draw $W_{t,d,n}\sim Categorical (f(\pi_{t,z}))$
      
Observe the fact that each time slice is just a separate LDA model.



(Fig. 3) Plate notation for the Dirichlet DTM, with 3 time slices.

Let's define the DTM class:

Note that the corpus should also be sliced by time. You shall feed your data using a list of list of lists, representing Time, Documents and Words. 

We will be fitting on the $nltk$ $inaugural$ corpus, and we'll asume that there is only a single slice of time (this is equivalent to the CTM).

For dummy experiments, use the following corpus:

```python
# docs_timeslice = [
#     [
#         ['aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa'],
#         ['bbb', 'bbb', 'bbb', 'bbb', 'bbb', 'bbb', 'bbb', 'bbb', 'bbb'],
#         ['bbb', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'bbb', 'bbb'],

#     ],
#     [
#         ['aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa'],
#         ['bbb', 'bbb', 'bbb', 'bbb', 'bbb', 'bbb', 'bbb', 'bbb'],
#         ['bbb', 'bbb', 'bbb', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'bbb', 'bbb'],

#     ],
#     [
#         ['aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa'],
#         ['bbb', 'aaa', 'aaa', 'aaa', 'aaa', 'aaa', 'bbb', 'bbb'],
#     ]
# ]
```

```python
class DTM_Topic_Dirichlet(object):
    
    def __init__(self, docs_time, K=2, alpha_scale=1, pi_scale=1):
        self.vocabulary = Vocabulary_Time(docs_time)
        
        self.collection = [[[float(self.vocabulary[w]) for w in d] for d in docs] for docs in docs_time]
        
        self.K = K
        self.T = len(self.collection) # Time slices
        self.collection = np.array(self.collection)
        
        self.V = len(self.vocabulary)
        self.D = [len(timeslice) for timeslice in self.collection]
        
        self.Nd = [[len(doc) for doc in timeslice] for timeslice in self.collection]
        
        self.alpha = np.ones(self.K) * alpha_scale

        self.pi_0_mu = np.ones(self.V) * pi_scale
        self.pi_cov = np.matmul(make_spd_matrix(self.V), np.eye(self.V))
        
        self.pi_0 = pm.MvNormal('pi_0', mu=self.pi_0_mu, tau=self.pi_cov)
        
        #Should the covariance matrix be the same for every Gaussian?

        self.pi_dict = {'pi_0': self.pi_0}
        
        for i in range(self.T - 1):
            self.pi_dict[f'pi_{i+1}'] = pm.MvNormal(f'pi_{i+1}', mu=self.pi_dict[f'pi_{i}'], tau=self.pi_cov)        

        self.pi = pm.Container(
            self.pi_dict
        )
        
        self._init_DTM()
        
    def _init_DTM(self):
        self.theta = pm.Container(
            [
                [
                    pm.CompletedDirichlet(
                        "theta_%s_%s" % (t, d),
                        pm.Dirichlet("ptheta_%s_%s" % (t, d), theta = self.alpha)
                    )
                    for d in range(self.D[t])
                ]
            for t in range(self.T) #time slices
            ]
        )
                
        self.z = pm.Container(
            [
                [
                    pm.Categorical(
                        "z_%i_%i" % (t, d),
                        p = self.theta[t][d],
                        size = self.Nd[t][d]
                    )
                    for d in range(self.D[t])    
                ]
                for t in range(self.T)
            ]
        )
        
        self.w = pm.Container(
            [
                pm.Categorical(
                    "w_%i_%i_%i" % (t, d, n),
                    pm.Lambda(
                        "sm_%i_%i_%i" % (t, d, n),
                        lambda z = self.z[t][d][n] : softmax(self.pi[f'pi_{t}'].value)
                    ),  
                    value = self.collection[t][d][n],
                    observed = True
                ) 
                for t in range(self.T) for d in range(self.D[t]) for n in range(self.Nd[t][d])
            ]
        )       

        self.model = pm.Model([self.pi, self.theta, self.z, self.w])
        self.mcmc = pm.MCMC(self.model)
#         self.mcmc = pm.MCMC([self.theta, self.phi, self.z, self.w])     
        
#     def fit(self, iter=40000, burn=15000, thin=10, tune_interval=2, tune_throughout=False):
    def fit(self, iter=40000, burn=15000):
        self.mcmc.sample(iter, burn)
```

```python
DTM_Model = DTM_Topic_Dirichlet(corpus)
```

```python
DTM_Model.fit(1000, 500)
```

```python
# To do: Fix the class (specifically the w and z variables)
#        Let every Gaussian in families have the same covariance matrix

class DTM_Topic_Gauss(object):
    
    def __init__(self, docs_time, K=2, alpha_scale=1, beta_scale=1):
        self.vocabulary = Vocabulary_Time(docs_time)
        self.collection = [[[self.vocabulary[w] for w in d] for d in docs] for docs in docs_time]
        self.K = K
        self.T = len(self.collection) # Time slices
                
        self.V = len(self.vocabulary)
        self.D = [len(c) for c in self.collection]
        
#         self.D = [2, 4, 3] # dummy values
        
#         self.Nd = [len(doc) for doc in self.collection]
#         self.Nd = [[3,4], [4,3,6,2], [6,4,3]]
        self.Nd = [[len(doc) for doc in c] for c in self.collection]
        
        print('self.Nd:', self.Nd)
        
        self.alpha_0_mu = np.ones(self.K) * alpha_scale
#         self.alpha_0_cov = np.matmul(make_spd_matrix(self.K), np.eye(self.K))
        self.alpha_0_cov = make_spd_matrix(self.K)          
        self.alpha_0 = pm.MvNormal('alpha_0', mu=self.alpha_0_mu, tau=self.alpha_0_cov)
        
        #Should the covariance matrix be the same for every Gaussian?
#         self.alpha_covs = [self.alpha_0_cov] + [np.matmul(make_spd_matrix(self.K), np.eye(self.K)) for i in range(self.T-1)]
        self.alpha_covs = [self.alpha_0_cov] + [make_spd_matrix(self.K) for i in range(self.T-1)]
        self.alpha_dict = {'alpha_0': self.alpha_0}
        
        for i in range(self.T - 1):
            self.alpha_dict[f'alpha_{i+1}'] = pm.MvNormal(f'alpha_{i+1}', mu=self.alpha_dict[f'alpha_{i}'], tau=self.alpha_covs[i+1])        

        self.alpha = pm.Container(
            self.alpha_dict
        )
        
        self.beta_0_mu = np.ones(self.V) * beta_scale
#         self.beta_0_cov = np.matmul(make_spd_matrix(self.V), np.eye(self.V))
        self.beta_0_cov = make_spd_matrix(self.V)
        self.beta_0 = pm.MvNormal('alpha_0', mu=self.beta_0_mu, tau=self.beta_0_cov)
        
        #Should the covariance matrix be the same for every Gaussian?
#         self.beta_covs = [self.beta_0_cov] + [np.matmul(make_spd_matrix(self.V), np.eye(self.V)) for i in range(self.T-1)]
        self.beta_covs = [self.beta_0_cov] + [make_spd_matrix(self.V) for i in range(self.T-1)]
        self.beta_dict = {'beta_0': self.beta_0}
        for i in range(self.T - 1):
            self.beta_dict[f'beta_{i+1}'] = pm.MvNormal(f'beta_{i+1}', mu=self.beta_dict[f'beta_{i}'], tau=self.beta_covs[i+1])        
        
        self.beta = pm.Container(
            self.beta_dict
        )
        
        self.eta_cov = np.matmul(make_spd_matrix(K), np.eye(K))
        
        self._init_DTM()
        
    def _init_DTM(self):
        self.eta = pm.Container(
            [
                pm.MvNormal(f'eta_{t}', mu=self.alpha_dict[f'alpha_{t}'].value, tau=self.eta_cov)
                for t in range(self.T)
            ]
        )
        
        self.z = pm.Container(
            [
                [
                    pm.Categorical(
                        "z_%i_%i" % (t, d),
                        p = softmax(self.eta[t].value),
                        size = self.Nd[d]
                    )
                    for d in range(self.D[t])    
                ]
                for t in range(self.T) # pt fiecare timeslice
            ]
        )
        
        self.w = pm.Container(
            [
                pm.Categorical(
                    "w_%i_%i_%i" % (t, d, i), #timeslice t, documentul d, cuvantul i
                    pm.Lambda(
                        "pi_%i_%i" % (t, i),
                        lambda z = self.z[t][d][i] : softmax(self.beta[f'beta_{t}'].value)
                    ),  
                    value = self.collection[t][d][i],
                    observed = True
                ) 
                for t in range(self.T) for d in range(self.D[t]) for i in range(self.Nd[t][d])
            ]
        )       
       
        self.model = pm.Model([self.eta, self.phi, self.z, self.w])
        self.mcmc = pm.MCMC(self.model, verbose=3)
#         self.mcmc = pm.MCMC([self.theta, self.phi, self.z, self.w])
        
    def get_parameters(self):
        return (self.V, self.D, self.alpha, self.beta, self.Nd)
    
    def print_parameters(self):
        print('V: %s \nD: %s \nalpha: %s \nbeta: %s \nNd: %s' % 
             self.get_parameters())        

        
#     def fit(self, iter=40000, burn=15000, thin=10, tune_interval=2, tune_throughout=False):
    def fit(self, iter=40000, burn=15000):
        self.mcmc.sample(iter, burn)
```
